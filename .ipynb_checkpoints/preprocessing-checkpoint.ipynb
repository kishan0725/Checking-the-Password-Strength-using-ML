```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
data.dropna(how='any',inplace=True)
```

```python
data.isna().sum()
```

```python
data.strength.value_counts()
```

```python
data1 = data.loc[(data.strength == '1') | (data.strength == '2') | (data.strength == '0'),:]
```

```python
data1.strength.value_counts()
```

```python
passwords_tuple=np.array(data1)
```

```python
import random
random.shuffle(passwords_tuple) #shuffling randomly for robustness
```

```python
y=[labels[1] for labels in passwords_tuple]
```

```python
y
```

```python
X=[labels[0] for labels in passwords_tuple]
```

```python
X
```

```python
def char_tokenize(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=char_tokenize)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
feature_names = vectorizer.get_feature_names()
feature_names
```

```python
print("Text: "+data1.iloc[0,0])
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=37)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
log_class=LogisticRegression(penalty='l2',multi_class='ovr')
log_class.fit(X_train,y_train)
```

```python
print(log_class.score(X_test,y_test))
```

```python
# Multinomial

clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train) #training
print(clf.score(X_test, y_test))
```

```python
X_predict=np.array(["abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```

```python
X_predict=np.array(["123abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```

```python
X_predict=np.array(["&*$abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```
